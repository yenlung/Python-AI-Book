## 1. 讀入基本套件

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. 一樣是有函式，就能互動!

```python
def bmi_cal(h, w):
    h = float(h)/100
    w = float(w)
    bmi = w/(h**2)
    message = f"你的 BMI 是 {bmi:.2f}。"
    return message
```

```python
bmi_cal(80, 170)
```

```python
import gradio as gr
```

## 4. Wep App 瞬間完成!

```python
iface = gr.Interface(bmi_cal, inputs=["number", "number"], outputs="text")
```

```python
iface.launch()
```

### 5. 可以做得更酷!

注意舊版的 Gradio 有分輸入和輸出的元件 (components), 但新版已不分了。意思是舊版的

    inp1 = gr.inputs.Number(...)

新版會簡化為:

    inp1 = gr.Number(...)

但新版兩種寫法都可以接受。

```python
inp1 = gr.Number(label="請輸入身高 (公分)")
inp2 = gr.Number(label="請輸入體重 (公斤)")
```

```python
iface = gr.Interface(bmi_cal, 
                    inputs = [inp1, inp2],
                    outputs = "text",
                    title="BMI 計算機", 
                    description="輸入你的身高體重, 幫你計算 BMI!"
)
 
```

```python
iface.launch()
```
